## Growing Strength
- Make a variable called strength, and set its initial value to 5.
- Print a message reporting the player's strength.
- Set up a while loop that runs until the player's strength increases to a value such as 10.
- Inside the while loop, print a message that reports the player's current strength.
- Inside the while loop, write a statement that increases the player's strength.
- Outside the while loop, print a message reporting that the player has grown too strong, and that - they have moved up to a new level of the game.
- __Bonus: Play around with different cutoff levels for the value of strength, and play around with different ways to increase the strength value within the while loop.__

````python
strength = 5
print("Your strength: " + str(strength))
while strength < 10:
    strength = strength + 2
    print("Your strength: " + str(strength))
print("You've grown too strong and will move up a level")
````

## Get the right response
Write a Python program that asks the user to respond by 'Y', 'y', 'yes', 'YES' or 'N', 'n', 'no', 'NO'. The function keeps on asking until the user enters the correct information. Use a list to store all possible answers.

````python
my_animals = ['dog', 'dog', 'dog']
print('cat' in my_animals)
print('dog' in my_animals)
````

````python
answers = ['Y', 'y', 'yes', 'YES', 'N', 'n', 'no', 'NO']
response = input('What is your response?')
while not response in answers:
    response = input('What is your response?')
    print("You said", response)
print("thanks for your response!")
````

## Guess the lucky number
Write a program so that it asks users to "guess the lucky number". If the correct number is guessed the program stops, otherwise it continues forever.

__Bonus__: Use `import random` and `random.randint(0, 10)` so you don't know what the number is

````python
import random
random_number = random.randint(0, 10)
print(random_number)
````

````python
number = int(input("Guess a number"))
correct_number = random.randint(0, 10)
while number != correct_number:
    print("False!")
    number = int(input("Guess a number"))
print("You guessed correctly!", correct_number)
````
